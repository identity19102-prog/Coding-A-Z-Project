```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

# ===== Python 공식 뉴스 URL =====
url = "https://www.python.org/blogs/"

# HTTP 요청
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# ===== 뉴스 제목, 링크, 날짜 추출 =====
headlines = []
articles = soup.select("ul.list-recent-posts li")  # 각 li가 하나의 글

for li in articles[:20]:  # 상위 20개
    a_tag = li.select_one("h3 a")
    date_tag = li.select_one("time")
    title = a_tag.get_text().strip()
    link = a_tag['href']
    date = date_tag.get_text().strip() if date_tag else ""
    headlines.append({"title": title, "link": link, "date": date})

# ===== CSV 저장 =====
today = datetime.today().strftime('%Y-%m-%d')
df = pd.DataFrame(headlines)
filename = f"python_org_news_{today}.csv"
df.to_csv(filename, index=False, encoding='utf-8-sig')

print(f"[{today}] Python.org 뉴스 {len(df)}개를 {filename}로 저장했습니다.")
print(df)
```
