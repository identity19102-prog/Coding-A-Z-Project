```python
import pandas as pd
import numpy as np

number = 1000

students = [f"Student{i+1}" for i in range(number)]

np.random.seed(42)
scores = {
    "Korean": np.clip(np.random.normal(75, 25, number).round(), 0, 100).astype(int),
    "English": np.clip(np.random.normal(75, 25, number).round(), 0, 100).astype(int),
    "Mathematics": np.clip(np.random.normal(75, 25, number).round(), 0, 100).astype(int),
    "Social": np.clip(np.random.normal(75, 25, number).round(), 0, 100).astype(int),
    "Science": np.clip(np.random.normal(75, 25, number).round(), 0, 100).astype(int)
}

df = pd.DataFrame(scores)
df.insert(0, "Student", students)

df["Average"] = df[["Korean","English","Mathematics","Social","Science"]].mean(axis=1)

def calc(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    elif avg >= 60:
        return "D"
    else:
        return "F"

df["Grade"] = df["Average"].apply(calc)

df.to_csv("students_grades.csv", index=False, encoding='utf-8-sig')

```
